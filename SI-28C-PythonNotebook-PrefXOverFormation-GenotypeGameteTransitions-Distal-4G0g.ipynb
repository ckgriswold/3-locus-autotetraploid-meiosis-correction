```python
import sympy as sp 
import itertools
a1,a2,b1,b2,a3,a4,b3,b2,b4,c1,c2,c3,c4,G,g = sp.symbols('a1,a2,b1,b2,a3,a4,b3,b2,b4,c1,c2,c3,c4,G,g')    

import copy
from sympy import *
from itertools import permutations 
from itertools import chain
from itertools import product
```

```python
A1,A2,B1,B2= sp.symbols('A1,A2,B1,B2') 
p = sp.Function('p') 
```

```python
geno = [[[i11,i12,i13],[i21,i22,i23],[i31,i32,i33],[i41,i42,i43]] \
                                                         for i11 in [A1,A2] for i12 in [B1,B2] for i13 in [G,g] \
                                                        for i21 in [A1,A2] for i22 in [B1,B2] for i23 in [G,g] \
                                                        for i31 in [A1,A2] for i32 in [B1,B2] for i33 in [G,g] \
                                                        for i41 in [A1,A2] for i42 in [B1,B2] for i43 in [G,g]]
```

```python
u_geno = [geno[0]]
for item in geno:
    flag = 0
    for perm in itertools.permutations(item):
        if list(perm) in u_geno: 
            flag = 1
    if flag == 0:
        u_geno.append(item)
```

```python
u_geno_sorted = []

for item in u_geno:
    Gg_arrange = [item[0][2],item[1][2],item[2][2],item[3][2]]
    num_G = Gg_arrange.count(G)
    if num_G == 4 or num_G == 0:
        u_geno_sorted.append(item)
    elif num_G == 3:
        index_G = [index for (index, item_if) in enumerate(Gg_arrange) if item_if == G]
        index_g = [index for (index, item_if) in enumerate(Gg_arrange) if item_if == g]
        tmp_1 = item[index_G[0]]
        tmp_2 = item[index_G[1]]
        tmp_3 = item[index_G[2]]
        
        tmp_4 = item[index_g[0]]
        
        u_geno_sorted.append([tmp_1,tmp_2,tmp_3,tmp_4])
    elif num_G == 2:
        index_G = [index for (index, item_if) in enumerate(Gg_arrange) if item_if == G]
        index_g = [index for (index, item_if) in enumerate(Gg_arrange) if item_if == g]
        tmp_1 = item[index_G[0]]
        tmp_2 = item[index_G[1]]
        
        tmp_3 = item[index_g[0]]
        tmp_4 = item[index_g[1]]
        
        u_geno_sorted.append([tmp_1,tmp_2,tmp_3,tmp_4])
    elif num_G == 1:
        index_G = [index for (index, item_if) in enumerate(Gg_arrange) if item_if == G]
        index_g = [index for (index, item_if) in enumerate(Gg_arrange) if item_if == g]
        tmp_1 = item[index_G[0]]
        
        tmp_2 = item[index_g[0]]
        tmp_3 = item[index_g[1]]
        tmp_4 = item[index_g[2]]
        
        u_geno_sorted.append([tmp_1,tmp_2,tmp_3,tmp_4])
    else:
        print('error')
```
